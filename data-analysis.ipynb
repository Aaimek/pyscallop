```python
import pandas as pd
import numpy as np
import helper
import matplotlib.pyplot as plt
from matplotlib.offsetbox import AnchoredText
from scipy.stats import norm
import seaborn as sns
```

## Afficher l'histograme des mesures d'une seule coquille

```python
df = pd.read_pickle('./output/data/all_contours.pkl')
df = df[df.state == 'GOOD']

lenghs = df.lengh * 100
lenghs = lenghs[lenghs > 10]
diameter = df.radius * 2

helper.draw_distrib(lenghs)
```

## Afficher l'évolution de la mesure de la coquille en fonction de la heuteur de la coquille dans l'image

```python
hauteur_image = []
mesures = []

for idex, value in df.iterrows():
    hauteur_image.append(value.center[1])
    mesures.append(value.lengh*100)

fig, ax = plt.subplots()
ax.set_xlabel("Hauteur dans l'image (piexels)")
ax.set_ylabel("Taille (cm)")
ax.set_title("Taille d'une même coquille")
ax.scatter(hauteur_image, mesures, marker='+')
ax.legend(['Mesures non corgiées'])
plt.savefig('./output/mesures_nc_scatter.jpg', dpi=1000)
```

## Faire une reg. linéaire puis corriger la mesure

```python
polynomial = np.polyfit(x=hauteur_image, y=mesures, deg=1)
a, b = polynomial[0], polynomial[1]

# créer un tablau lingle avec la reglin
line = []
for hauteur in hauteur_image:
    lin = a*hauteur + b
    line.append(lin)

# mesures corrigés: mesures orginiale moins a*x (on garde l'ordonée a l'origine)
mesures_corrigés = mesures[:]
for i in range(len(line)):
    mesures_corrigés[i] = mesures[i] - a*hauteur_image[i]


fig, ax = plt.subplots()
width_inch = 7
height_inch = 0.75 * width_inch
fig.set_size_inches(width_inch, height_inch)
text = "Coeff dirr: " + str(round(a, 2)) + "\n"
text += "Ord orig.: " + str(round(b, 2))
at = AnchoredText(text, prop=dict(size=10), frameon=True, loc='upper left')
at.patch.set_boxstyle("round,pad=0.,rounding_size=0.2")
ax.add_artist(at)
ax.scatter(hauteur_image, mesures, marker='+')
ax.plot(hauteur_image, line, color='orange')
ax.set_title("Détermination du défaut d'angle de la caméra")
ax.set_xlabel("Hauteur dans l'image (piexels)")
ax.set_ylabel("Mesure (cm)")
ax.legend(['Régression linéaire', 'Mesures non corrigés'])
plt.savefig('./output/determination_angle_camera.jpg', dpi=1000)
```

## Afficher les mesures corrigées

```python
fig, ax = plt.subplots()
width_inch = 7
height_inch = 0.75 * width_inch
fig.set_size_inches(width_inch, height_inch)

ax.scatter(hauteur_image, mesures_corrigés, marker='+')
ax.legend(['mesures corrigées'])
ax.set_xlabel("Hauteur dans l'image (pixels)")
ax.set_ylabel("Mesure (cm)")
ax.set_title("Mesures corrigées")
plt.savefig("./output/mesures_corigées_scatter.jpg", dpi=1000)
```

## Comparer l'histograme des mesures corrigées vs non corrigées

```python
raw_mean, raw_std = norm.fit(mesures)
corr_mean, corr_std = norm.fit(mesures_corrigés)

fig, ax = plt.subplots()
# text = "Moyenne: " + str(round(mu, 2)) + " pixels\n"
hist_text = "Ecart-type non corrigé: " + str(round(raw_std, 2)) + " cm\n"
hist_text += "Ecart-type corrigé: " + str(round(corr_std, 2)) + " cm\n"
hist_text += "Nombre de mesures: " + str(len(mesures))
at = AnchoredText(hist_text, prop=dict(size=10), frameon=True, loc='upper left')
at.patch.set_boxstyle("round,pad=0.,rounding_size=0.2")
ax.add_artist(at)

width_inch = 7
height_inch = 0.75 * width_inch
fig.set_size_inches(width_inch, height_inch)

ax.hist(mesures, bins='auto', alpha=0.5)
ax.hist(mesures_corrigés, alpha=0.5)
ax.set_title("Histogramme des mesures d'une même coquille")
ax.set_xlabel("Taille (cm)")
ax.set_ylabel("Nombre de mesures")
# ax.axvline(mu, color='k', linestyle='dashed', linewidth=1)
ax.legend(['Non corrigé', 'Corrigé'])
plt.savefig('./output/hist_corrigé_vs_noncorr.jpg', dpi=1000)
```

## Histograme + distribution normale ajustée des mesures corrigés

```python
fig, ax = plt.subplots()
txt = "Moyenne: " + str(round(corr_mean, 2)) + 'cm\n'
txt += "Ecart-type: " + str(round(corr_std, 2)) + 'cm\n'
txt += "Nombre de mesures: " + str(len(mesures_corrigés))
at = AnchoredText(txt, prop=dict(size=10), frameon=True, loc='upper left')
at.patch.set_boxstyle("round,pad=0.,rounding_size=0.2")
ax.add_artist(at)
ax.set_title("Histogramme des mesures, corrigé par reg. lin.")

# Taille
width_inch = 7
height_inch = 0.75 * width_inch
fig.set_size_inches(width_inch, height_inch)

# Histogramme
ax.hist(mesures_corrigés, color='orange', density=True)
ax.axvline(corr_mean, color='k', linestyle='dashed', linewidth=1)
ax.set_xlabel("Taille (cm)")
ax.set_ylabel("Nombre de mesures")

# Fit de distrubution normale
domain = np.linspace(np.min(mesures_corrigés), np.max(mesures_corrigés))
best_fit_line = norm.pdf(domain, corr_mean, corr_std)
ax.plot(domain, best_fit_line)

# Legende
ax.legend(['Moyenne', 'Distribution normale', 'Mesures corrigées'])

# Save
plt.savefig('./output/hitogramme_mesures_corrigés.jpg', dpi=1000)
```
